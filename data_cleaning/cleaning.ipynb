## Working with null columns

```python
columns = ["age","experience_years","daily_work_hours","sleep_hours",
           "caffeine_intake","bugs_per_day","screen_time","meetings_per_day","commits_per_day","exercise_hours","stress_level"]

for i in range(len(columns)):
    col = columns[i]
    data[col] = data[col].replace("null", pd.NA)
    data[col] = data[col].fillna(data[col].median())
```

```python
data["screen_time"] = data["screen_time"].replace(0, data["screen_time"].median())
```

```python
data["burnout_level"] = data["burnout_level"].replace("<null>", pd.NA)
data["burnout_level"] = data["burnout_level"].fillna(data["burnout_level"].mode()[0])
```

```python
data.isnull().sum()
```

## Correcting the datatype

```python
data.head()
columns = ["age","experience_years","caffeine_intake","bugs_per_day","meetings_per_day","commits_per_day","stress_level"]

for i in range(len(columns)):
    col = columns[i]
    data[col] = data[col].astype(np.int32)
```

## Correcting column names

```python
data.columns = data.columns.str.capitalize()
```

```python
data.head()
```

## Age Group Column Addition

```python
def aging(x):
    if x <= 25:
        return "20-25"
    elif x <= 30:
        return "26-30"
    elif x <= 35:
        return "31-35"
    elif x <= 40:
        return "36-40"
    else:
        return "41-45"

data["Age_Group"] = data["Age"].apply(aging)
```
